# GroupBy

```python
df.groupby(["Type 1","Type 2"]).count()
```

```python
new_df = pd.DataFrame(columns=df.columns)
for df_ in pd.read_csv('modified.txt',sep='\t',chunksize=5):
    res = df_.groupby(["Type 1"]).count()
    new_df = pd.concat([new_df,res])
```
